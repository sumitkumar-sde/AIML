```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Image Addition

```python
img1 = cv2.imread('images/lenna.png')
img2 = cv2.imread('images/sun.png')
img1.shape, img2.shape
```

```python
add_img = cv2.add(img1, img2)
cv2.imshow('Added Image', add_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Using MatplotLib


```python
imgM1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
imgM2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

added = cv2.cvtColor(add_img, cv2.COLOR_BGR2RGB)

#show
plt.figure(figsize=(16,9))
plt.subplot(1,3,1)
plt.imshow(imgM1)
plt.subplot(1,3,2)
plt.imshow(imgM2)
plt.subplot(1,3,3)
plt.imshow(added)
```

# Substraction

```python
sub_img = cv2.subtract(imgM1, imgM2)
#show
plt.figure(figsize=(16,9))
plt.subplot(1,3,1)
plt.imshow(imgM1)
plt.subplot(1,3,2)
plt.imshow(imgM2)
plt.subplot(1,3,3)
plt.imshow(sub_img)
```

# Add weighted

```python
alpha = 0.25 #Transparency Factor

blended_img = cv2.addWeighted(imgM1, alpha, imgM2, 1-alpha, 0)

plt.figure(figsize=(16,9))
plt.subplot(1,3,1)
plt.imshow(imgM1)
plt.subplot(1,3,2)
plt.imshow(imgM2)
plt.subplot(1,3,3)
plt.imshow(blended_img)
```

# Multiplication

```python
bright_img = cv2.multiply(imgM2, 3.19)
darken_img = cv2.multiply(imgM2, 0.19)
plt.figure(figsize=(16,9))
plt.subplot(1,3,1)
plt.imshow(bright_img)
plt.subplot(1,3,2)
plt.imshow(darken_img)
```

# Bitwise And & OR


```python
input1 = cv2.imread('images/input1.png')
input2 = cv2.imread('images/input2.png')

dest_and = cv2.bitwise_and(input1, input2, mask = None)
dest_or = cv2.bitwise_or(input1, input2, mask = None)
dest_xor = cv2.bitwise_xor(input1, input2, mask = None)

plt.figure(figsize=(16,9))
plt.title('AND')
plt.subplot(1,9,1)
plt.imshow(input1)
plt.subplot(1,9,2)
plt.imshow(input2)
plt.subplot(1,9,3)
plt.imshow(dest_and)

plt.title('OR')
plt.subplot(2,9,1)
plt.imshow(input1)
plt.subplot(2,9,2)
plt.imshow(input2)
plt.subplot(2,9,3)
plt.imshow(dest_or)

plt.title('XOR')
plt.subplot(3,9,1)
plt.imshow(input1)
plt.subplot(3,9,2)
plt.imshow(input2)
plt.subplot(3,9,3)
plt.imshow(dest_xor)
```
