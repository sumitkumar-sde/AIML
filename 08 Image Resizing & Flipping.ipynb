```python
import cv2

img = cv2.imread('images/lenna.png')

flip_Horizontal = cv2.flip(img, 0)
flip_Vertical = cv2.flip(img, 1)
flip_Diagonal = cv2.flip(img, -1)

cv2.imshow('Fliped Image', flip_Vertical)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
rgbF_H = cv2.cvtColor(flip_Horizontal, cv2.COLOR_BGR2RGB)
rgbF_V = cv2.cvtColor(flip_Vertical, cv2.COLOR_BGR2RGB)
rgbF_D = cv2.cvtColor(flip_Diagonal, cv2.COLOR_BGR2RGB)
rgb_Org = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20,17))
plt.subplot(1,4,1)
plt.title('Original')
plt.imshow(rgb_Org)

plt.subplot(1,4,2)
plt.title('Horizontal')
plt.imshow(rgbF_H)

plt.subplot(1,4,3)
plt.title('Vertical')
plt.imshow(rgbF_V)

plt.subplot(1,4,4)
plt.title('Diagonal')
plt.imshow(rgbF_D)
```
