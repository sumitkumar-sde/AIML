```python
df_c['AverageRun/Match'] = df_c['runs']/df_c['matches']
df_c
```

```python
df_c.columns
```

```python
#Create the mask
df_c['AverageRun/Match'] > 50
```

```python
# Apply the mask to DataFrame
df_c[df_c['AverageRun/Match'] > 50]
```
