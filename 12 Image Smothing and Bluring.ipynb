```python
import cv2
import numpy as np

#Load the image
img = cv2.imread('images/lenna.png')

kernel = np.ones((7,7), np.float32)/49

#Apply to the Kernel
smotthed_image = cv2.filter2D(img, -10, kernel)

cv2.imshow('Original Image', img)
cv2.imshow('Smotthed Image', smotthed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Gaussian Blur

```python
import matplotlib.pyplot as plt

gray = cv2.imread('images/lenna.png', 0)
kernel_size = 5
kernel = cv2.getGaussianKernel(kernel_size, 1.25)
blured_img = cv2.filter2D(gray, -1, kernel)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(gray, cmap='gray')
plt.subplot(1,2,2)
plt.title('Gaussian B Image')
plt.imshow(blured_img, cmap='gray')
```
