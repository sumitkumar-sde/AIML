# Image Shearing

```python
shear_x = 0.25
shear_y = 0.45

#Create Transformation matrix
M = np.array([[1,shear_x, 0],[shear_y, 1, 0],[0,0,1]])

row = new_image.shape[0]
col = new_image.shape[1]

T = np.float32([[1,0,0],[0,1,50]])

#we use wrapAffine to Transfrom the image using the matrix, T
img_translation = cv2.warpAffine(new_image, T, (int(col*1.5), int(row*1.5)))
sheard_img = cv2.warpPerspective(img_translation, M, (int(col*1.5), int(row*1.5)))

plt.imshow(sheard_img)
```
