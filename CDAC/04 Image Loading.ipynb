```python
import matplotlib.pyplot as plt
img_plt = plt.imread('images/lenna.png')

img_plt.shape
```

```python
mat1 = img_plt[0:512,0:512,0:1]
mat2 = img_plt[0:512,0:512,1:2]
mat3 = img_plt[0:512,0:512,2:3]
plt.imshow(mat3)
```

```python
r=img_plt[:,:,0]
g=img_plt[:,:,1]
b=img_plt[:,:,2]
plt.imshow(b)
r.shape, g.shape, b.shape
```

```python
plt.subplot(1,3,1)
plt.imshow(r,cmap='Reds')
plt.subplot(1,3,2)
plt.imshow(g,cmap='Greens')
plt.subplot(1,3,3)
plt.imshow(b,cmap='Blues')
```

```python
new_img = (r+g+b)/2
```

```python
plt.imshow(new_img, cmap='gray')
```

```python
import cv2
img1 = cv2.imread('images/lenna.png')
plt.subplot(1,4,1)
plt.title('Image Read using CV2')
plt.imshow(img1)
plt.subplot(1,4,2)
plt.title('Image Read using PLT')
plt.imshow(img_plt)
plt.subplot(1,4,3)
plt.title('Image Read using CV2 convert')
plt.imshow(img1)
plt.subplot(1,4,4)
plt.title('Image Read using PLT convert')
plt.imshow(img_plt)
```

```python
import cv2
img1 = cv2.imread('images/lenna.png')
plt.subplot(1,2,1)
plt.title('Image Read using CV2')
img2= cv2.cvtColor(img1, cv2.COLOR_BGR2XYZ)
plt.imshow(img2)
plt.subplot(1,2,2)
plt.title('Image Read using PLT')
plt.imshow(img_plt)
```
