```python
# Create the Datasets
```

```python
dataset = [
    ['Apple', 'Beer', 'Rice', 'Chicken'],
    ['Apple', 'Beer', 'Rice'],
    ['Apple', 'Beer'],
    ['Apple', 'Pear'],
    ['Milk', 'Beer', 'Rice', 'Chicken'],
    ['Milk', 'Beer', 'Rice'],
    ['Milk', 'Beer'],
    ['Milk', 'Pear']
]
dataset
```

```python
# convert to structure formate
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
encoded = te.fit_transform(dataset)
encoded
```

```python
te.columns_
```

```python
df = pd.DataFrame(encoded, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# 1. Find frequent item set
freq_itemset = apriori(df, min_support=0.25, use_colnames=True)
freq_itemset
```

```python
#rule = association_rules(freq_itemset, metric='lift', min_threshold=0.5)

rule = association_rules(freq_itemset, metric='confidence', min_threshold=0.5)
rule
```

```python
rules = rule[['antecedents', 'consequents', 'support', 'confidence']]
rules
```

### Sort the rules

```python
rules.sort_values(by=['confidence', 'support'], ascending=False)
```

#### Filter the Rules

```python
nrules = rules[rules['confidence'] > 0.5]
nrules
```

```python
# Recomandation System
```

```python
nrules[nrules['antecedents'] == {'Rice'}]['consequents']
```

```python
nrules[nrules['antecedents'] == {'Apple'}]['consequents']
```
