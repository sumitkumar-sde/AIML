# Sobel 

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
#import image as Gray Image
img = cv2.imread('images/lenna.png', 0)

#Sobel parameter: ksize, dx, dy, ddepth
sobelx = cv2.Sobel(img, cv2.CV_64F, 1,0, ksize=3)
sobely = cv2.Sobel(img, cv2.CV_64F, 0,1, ksize=3)

combined_sobel = cv2.add(sobelx, sobely)


Nsobelx = cv2.bitwise_not(combined_sobel)

plt.figure(figsize=(16,9))
plt.subplot(131)
plt.title('SobelX')
plt.imshow(sobelx , cmap = 'gray')

plt.subplot(132)
plt.title('SobelY')
plt.imshow(abs(sobely), cmap = 'gray')

plt.subplot(133)
plt.title('SobelX')
plt.imshow(abs(combined_sobel), cmap = 'gray')
```
